这里要下载的是NASA的NEX-GDDP-CMIP6，根据网页介绍，直接使用wget下载即可。

使用wget需要先找到URL，找到URL前，有必要了解下数据基本情况：根据[NEX-GDDP-CMIP6-Tech_Note](https://www.nccs.nasa.gov/sites/default/files/NEX-GDDP-CMIP6-Tech_Note.pdf)介绍：“the dataset compiles climate projections from thirty-five CMIP6 GCMs”，所以在[AMES/NEX/GDDP-CMIP6](https://ds.nccs.nasa.gov/thredds2/catalog/AMES/NEX/GDDP-CMIP6/catalog.html)的目录下面有35个文件夹。每个下面分别对应不同的场景，分别是未来情景的 ssp585 和 ssp585，以及历史的 historical 文件夹，它们各自下面又是每个变量一个文件夹，其中每年对应一个netcdf文件。每个单独的文件大约200+M，总共的数据约18T。

所以一次性下载完还是有点麻烦，可以尝试根据自己的地区，下载下固定范围的netcdf subset数据，再进行自己后续的处理。

下面先以一个URL的获取为例，看看如何在python下使用wget下载数据。

按照前面的介绍，选中一个nc文件后，点进去，可以看到类似如下页面：

![](QQ截图20220123215125.png)

这里选择netcdfsubset，进入后在页面下部可以看到URL，这里选择自己想要下载的区域，比如选择CONUS范围，则对应会生成下面的URL：

https://ds.nccs.nasa.gov/thredds2/ncss/AMES/NEX/GDDP-CMIP6/ACCESS-CM2/ssp585/r1i1p1f1/tasmin/tasmin_day_ACCESS-CM2_ssp585_r1i1p1f1_gn_2015.nc?var=tasmin&north=51&west=234&east=294&south=23&disableProjSubset=on&horizStride=1&time_start=2015-01-01T12%3A00%3A00Z&time_end=2015-12-31T12%3A00%3A00Z&timeStride=1

那么下面就试一试下载该URL对应的数据，用下str的format把参数提出来，方便后续的批量下载

```python
import wget
url_pattern="https://ds.nccs.nasa.gov/thredds2/ncss/AMES/NEX/GDDP-CMIP6/{gcm}/{scenario}/r1i1p1f1/{var}/{var}_day_{gcm}_{scenario}_r1i1p1f1_gn_{year}.nc?var={var}&north={north}&west={west}&east={east}&south={south}&disableProjSubset=on&horizStride=1&time_start={year}-01-01T12%3A00%3A00Z&time_end={year}-12-31T12%3A00%3A00Z&timeStride=1"
url=url_pattern.format(gcm='ACCESS-CM2',scenario='ssp585',year=2015,var='tasmin', north=51,west=234,east=294,south=23)
print(url)
wget.download(url)
```
